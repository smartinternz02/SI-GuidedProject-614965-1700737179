```python
from transformers import PegasusForConditionalGeneration, AutoTokenizer
import torch
from bs4 import BeautifulSoup
import requests
```

```python
URL="https://medium.com/inside-machine-learning/what-is-a-transformer-d07dd1fbec04"
```

```python
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
results = soup.find_all(['h1', 'p'])
text = [result.text for result in results]
ARTICLE = ' '.join(text)
```

```python
ARTICLE
```

```python
ARTICLE = ARTICLE.replace('.', '.<eos>')
ARTICLE = ARTICLE.replace('?', '?<eos>')
ARTICLE = ARTICLE.replace('!', '!<eos>')
```

```python
sentences = ARTICLE.split('<eos>')
```

```python
sentences
```

```python
max_chunk = 400
current_chunk = 0
chunks = []
for sentence in sentences:
    if len(chunks) == current_chunk + 1:
        if len(chunks[current_chunk]) + len(sentence.split(' ')) <= max_chunk:
            chunks[current_chunk].extend(sentence.split(' '))
        else:
            current_chunk += 1
            chunks.append(sentence.split(' '))
    else:
        print(current_chunk)
        chunks.append(sentence.split(' '))
```

```python
chunks
```

```python
for chunk_id in range(len(chunks)):
    chunks[chunk_id] = ' '.join(chunks[chunk_id])
```

```python
model_name ='google/pegasus-xsum'
device = 'cuda' if torch.cuda.is_available() else 'cpu'
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
model = PegasusForConditionalGeneration.from_pretrained(model_name).to(device)
```

```python
print(model)
```

```python
total_params = sum(p.numel() for p in model.parameters())
print(f"Total parameters: {total_params}")
```

```python
batch = tokenizer(chunks, truncation=True, padding='longest', return_tensors="pt").to(device)
```

```python
batch
```

```python
translated = model.generate(**batch)
```

```python
translated
```

```python
tgt_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
```

```python
tgt_text
```

```python
text = ' '.join([summ for summ in tgt_text])
```

```python
from datasets import load_dataset

dataset = load_dataset("xsum")

print(f"Features in xsum: {dataset['train'].column_names}")
```

```python
sample = dataset["train"][1]
print(f"""
Article (excerpt of 500 characters, total length: {len(sample["document"])}):
""")
print(sample["document"][:500])
print(f'\nSummary (length: {len(sample["summary"])}):')
print(sample["summary"])
```

```python
sample_text = dataset["train"][1]["document"][:1000]

# We'll collect the generated summaries of each model in a dictionary
summaries = {}
```

```python
batch = tokenizer(sample_text, truncation=True, padding='longest', return_tensors="pt").to(device)
```

```python
translated = model.generate(**batch)
```

```python
tgt_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
```

```python
tgt_text
```

```python
summaries["pegasus"]=tgt_text[0]
```

```python
summaries
```

```python
import pandas as pd
```

```python
rouge_metric = load_metric('rouge')
```

```python
summaries
```

```python
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]

reference = dataset['train'][1]['document']

records = []

for model_name in summaries:
    rouge_metric.add(prediction = summaries[model_name], reference = reference )
    score = rouge_metric.compute()
    rouge_dict = dict((rn, score[rn].mid.fmeasure ) for rn in rouge_names )
    print('rouge_dict ', rouge_dict )
    records.append(rouge_dict)

pd.DataFrame.from_records(records, index = summaries.keys() )
```
